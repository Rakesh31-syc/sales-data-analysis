```python
# Import necessary libraries
import sys
import os

# Add the scripts directory to the Python path
sys.path.append(os.path.abspath('../scripts'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from data_cleaning import load_data, clean_data
from eda import plot_sales_trends
from forecasting import forecast_sales

# Load and Clean Data
sales_data = load_data('../data/sales_data.csv')
cleaned_data = clean_data(sales_data)
cleaned_data.to_csv('../data/cleaned_sales_data.csv', index=False)
cleaned_data.head()
```

```python
# Plot sales trends
plt.figure(figsize=(12, 6))
plot_sales_trends(cleaned_data)
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()  # Show the plot
```

```python
# Calculate revenue by product
revenue_by_product = cleaned_data.groupby('product_id')['revenue'].sum().reset_index()

# Plot revenue by product
plt.figure(figsize=(10, 5))
sns.barplot(data=revenue_by_product, x='product_id', y='revenue')
plt.title('Revenue by Product')
plt.xlabel('Product ID')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()  # Show the plot
```

```python
# Calculate revenue by marketing campaign
revenue_by_campaign = cleaned_data.groupby('marketing_campaign')['revenue'].sum().reset_index()

# Plot revenue by marketing campaign
plt.figure(figsize=(10, 5))
sns.barplot(data=revenue_by_campaign, x='marketing_campaign', y='revenue')
plt.title('Revenue by Marketing Campaign')
plt.xlabel('Marketing Campaign')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()  # Show the plot
```

```python
# Forecast future sales
forecast = forecast_sales(cleaned_data)

# Display the forecasted values
print("Forecasted Sales for the Next 12 Periods:")
print(forecast)
```
